```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("whitegrid")
```

```python
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
iris = sns.load_dataset("iris")
```

```python
tips.head()
```

## Distribution Plots

```python
sns.displot(tips["total_bill"])
plt.show()
```

```python
sns.displot(tips["total_bill"], bins=20)
plt.show()
```

```python
sns.histplot(tips["total_bill"], bins=20)
plt.show()
```

```python
plt.figure(figsize=(12,9))
sns.kdeplot(tips["total_bill"])
```

```python
plt.figure(figsize=(12,9))
sns.pairplot(tips, hue="day")
```

```python
sns.jointplot(data=tips, x="total_bill",y="tip", height=10)
```

```python
sns.jointplot(data=tips, x="total_bill",y="tip", kind="hex")
```

```python
plt.figure(figsize=(12,9))
sns.jointplot(data=tips, x="total_bill",y="tip",hue="sex")
plt.show()
```

```python
plt.figure(figsize=(12,9))
sns.jointplot(data=tips, x="total_bill",y="tip", kind="reg")
```

```python
sns.rugplot(tips["total_bill"])
```

```python
plt.figure(figsize=(12,9))
sns.regplot(data=tips, x="total_bill",y="tip")
```

## Categorical Plots

```python
plt.figure(figsize=(12,9))
sns.barplot(data=tips,x="sex", y="total_bill")
```

```python
plt.figure(figsize=(12,9))
sns.countplot(data=tips,x="sex")
```

```python
#quartiles
plt.figure(figsize=(12,9))
sns.boxplot(data=tips,x="day", y="total_bill", hue="smoker")
```

```python
plt.figure(figsize=(15,12))
sns.violinplot(data=tips,x="day", y="total_bill", hue="smoker")
```

```python
plt.figure(figsize=(15,12))
sns.violinplot(data=tips,x="day", y="total_bill", hue="sex",split=True)
```

```python
plt.figure(figsize=(12,9))
sns.stripplot(data=tips,x="day", y="total_bill", hue="smoker")
```

```python
plt.figure(figsize=(12,9))
sns.swarmplot(data=tips,x="day", y="total_bill", hue="smoker")
```

```python
plt.figure(figsize=(12,9))
sns.catplot(data=tips,x="day", y="total_bill", hue="smoker", kind="point")
```

## Grids

```python
iris.head()
```

```python
sns.pairplot(iris)
```

```python
pg = sns.PairGrid(iris)
pg.map_diag(sns.histplot)
pg.map_upper(sns.scatterplot)
pg.map_lower(sns.kdeplot)
```

```python
fg = sns.FacetGrid(data=tips, col="time", row="smoker", palette="inferno")
fg.map(sns.histplot,"total_bill", "tip")
```

## Regression Plot

```python
plt.figure(figsize=(12,9))
sns.lmplot(data=tips, x="total_bill",y="tip",hue="sex", palette="coolwarm")
```

```python
sns.lmplot(data=tips, x="total_bill",y="tip", row="time", col="sex",height=10)
```
