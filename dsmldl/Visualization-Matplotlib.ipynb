```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
plt.plot(x,y,"r:")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("X-Y")
plt.show()
```

```python
#subplot
plt.subplot(1,2,1)
plt.plot(x,y,"red")
plt.subplot(1,2,2)
plt.plot(x+2,y+10,"green")
```

```python
#object oriented method
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.set_xlabel("***X***")
ax.plot(x,y)
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.2,0.3,0.4,0.5])
ax1.plot(x,y)
ax2.plot(y,x)
```

```python
fig,axes = plt.subplots(nrows=3,ncols=2)
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=2,ncols=3)
plt.tight_layout()
for axe in axes:
    if axes.ndim > 1:
        for ax in axe:
            ax.plot(x,y)
            ax.set_title("***")
    else:    
        axe.plot(x,y)
```

```python
#figure size and dpi
plt.figure(figsize=(12,9), dpi=75)
plt.plot(x,y)
plt.savefig("plt.png")
```

```python
plt.figure(figsize=(12,9), dpi=75)
plt.plot(x,y, label="x y")
plt.plot(x*2,y*2, label="x*2 y2")
plt.legend(loc="best")
```

```python
plt.figure(figsize=(12,9), dpi=75)
plt.plot(x,y, label="x y", color="green",lw=4,linestyle="-.")
plt.plot(x*2,y*2, label="x*2 y2", color="#AC5542",linewidth=10,alpha=0.5, marker="o", markersize=15, markerfacecolor="blue")
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.set_xlim([0,6])
ax1.set_ylim([1,4])
ax1.plot(y,x)
```
