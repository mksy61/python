```python
#import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("csv/fake_reg.csv")
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
from sklearn.model_selection import train_test_split
X = df[["feature1", "feature2"]].values
y = df["price"].values
```

```python
X
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=61)
```

```python
X_train.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([Dense(4,activation="relu"),
                    Dense(2,activation="relu"),
                    Dense(1)])
```

```python
model = Sequential()
model.add(Dense(4,activation="relu"))
model.add(Dense(4,activation="relu"))
model.add(Dense(4,activation="relu"))
model.add(Dense(1))
model.compile(optimizer="rmsprop", loss="mse")
```

```python
model.fit(x=X_train,y=y_train,epochs=600)
```

```python
loss_df = pd.DataFrame(model.history.history)
```

```python
loss_df.plot.line(figsize=(12,9))
```

```python
model.evaluate(x=X_test,y=y_test)
```

```python
model.evaluate(x=X_train,y=y_train)
```

```python
pred = model.predict(X_test)
```

```python
pred = pd.DataFrame(pred, columns=["Predictions"])
```

```python
pred
```

```python
pred_df = pd.DataFrame(y_test, columns=["Test True Y"])
```

```python
pred_df
```

```python
pred_df = pred_df.join(pred)
```

```python
pred_df
```

```python
plt.figure(figsize=(12,10))
sns.scatterplot(data=pred_df, x=pred_df["Test True Y"], y=pred_df["Predictions"])
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(pred_df["Test True Y"], pred_df["Predictions"])
```

```python
mean_squared_error(pred_df["Test True Y"], pred_df["Predictions"])
```
